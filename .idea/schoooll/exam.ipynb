```python
print('hello')

import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4])
plt.ylabel('number')#y축 이름입니다.
plt.show()#pylplot 노출
```

```python
plt.plot([1,2,3,4],[1,4,9,16])#x y 좌표를 잇는 그래프
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20]) #[xmin,xmax,ymin,ymax]
plt.show()

#해당 좌표에 점찍기입니다.
#axis의 경우에는 x좌표와 y좌표의 최소값과 최대값을 나타냅니다.
```

```python
t= np.arange(0.,5.,0.2)#0.0 0.2 0.4 ~5.0까지 점이 찍힙니다.
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')#x y좌표 t t 점선 // t t**2 네모점 // t t**3 세모
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values = [1,10,100]
plt.figure(figsize=(9,3)) #가로 , 세로 인치 단위입니다. 가로 9인치 세로 3인치짜리 플롯을 만듭니다.
plt.subplot(1,3,1) #1행 3열 중 1 # 첫번째 열의 서브플롯입니다.
plt.bar(names,values) #바형태 플롯 그룹 3개 세로 1 10 100짜리
plt.subplot(1,3,2) #1행 3열 중 2
plt.scatter(names,values)
plt.subplot(1,3,3)#1행 3열 중 3
plt.plot(names,values)
plt.suptitle('Categorical Plotting') #플롯의 이름..
plt.show()
```

```python
t = np.arange(0.0,2.0,0.01)
s=1+np.sin(2*np.pi*t)
fig, ax = plt.subplots() #fig와 ax라는 subplot
ax.plot(t,s)
ax.set(xlabel ='time (s)',ylabel='voltage(mV)',title = 'Sample Plot') #x축 y축  제목 ...
ax.grid()#그리드 형식으로
fig.savefig("test.png")
plt.show()
```

```python
x1 = np.linspace(0.0,5.0) #균일 간격 점 생성
x2 = np.linspace(0.0,2.0) #균일 
y1 = np.cos(2*np.pi*x1)*np.exp(-x1) #np.exp는 지수함수형태로 변환합니다.
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)# 2행 1열 중 첫번째 서브플롯
plt.plot(x1,y1,'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.subplot(2,1,2)# 2행 1열 두번째 서브플롯
plt.plot(x2,y2,'.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.show()
```
