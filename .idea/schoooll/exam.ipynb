```python
print('hello')

import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4])
plt.ylabel('number')#y축 이름입니다.
plt.show()#pylplot 노출
```

```python
plt.plot([1,2,3,4],[1,4,9,16])#x y 좌표를 잇는 그래프
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20]) #[xmin,xmax,ymin,ymax]
plt.show()

#해당 좌표에 점찍기입니다.
#axis의 경우에는 x좌표와 y좌표의 최소값과 최대값을 나타냅니다.
```

```python
t= np.arange(0.,5.,0.2)#0.0 0.2 0.4 ~5.0까지 점이 찍힙니다.
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')#x y좌표 t t 점선 // t t**2 네모점 // t t**3 세모
plt.show()
```
